# Transactional Data Analysis

## Dataset Features:
1. **Import/Export Indicator**: Indicates whether a transaction was an import or export.
2. **Product Categories**: Categorizes transactions based on the type of product involved.
3. **Shipping Method**: Method used to ship the product (e.g., air, sea, ground shipping).
4. **Supplier**: Information about the suppliers involved in the transactions.
5. **Payment Terms**: Terms agreed upon between the buyer and supplier (e.g., net 30 days, prepaid).
6. **Number of Transactions**: Records the number of transactions for different products, suppliers, and categories.

## Data Visualization Techniques:
- **Bar Plots**: Display the count of transactions by different categories (product, supplier, etc.).
- **Pie Charts**: Show the distribution of transactions by import/export status, shipping methods, and payment terms.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Display first few rows to inspect the dataset
print(data.head())

# Check if the necessary columns exist (replace with actual column names)
if 'Country' in data.columns and 'Product_Category' in data.columns:
    
    # Create a contingency table (cross-tabulation) of the two categorical variables
    contingency_table = pd.crosstab(data['Country'], data['Product_Category'])
    
    # Perform the Chi-Square test
    chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

    # Print the results
    print("\nChi-Square Test Results:")
    print(f"Chi-Square Statistic: {chi2_stat}")
    print(f"P-value: {p_value}")
    print(f"Degrees of Freedom: {dof}")
    print(f"Expected Frequencies Table:\n{expected}")

    # Interpretation of the p-value
    alpha = 0.05  # significance level
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant association between 'Country' and 'Product_Category'.")
    else:
        print("Fail to reject the null hypothesis: No significant association between 'Country' and 'Product_Category'.")

    # 1. Heatmap of the Contingency Table
    plt.figure(figsize=(12, 6))
    sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues', cbar=True, linewidths=0.5)
    plt.title('Heatmap of Contingency Table: Country vs Product_Category', fontsize=16)
    plt.xlabel('Product Category', fontsize=12)
    plt.ylabel('Country', fontsize=12)
    plt.show()

    # 2. Clustered Bar Chart
    plt.figure(figsize=(12, 6))
    contingency_table.plot(kind='bar', stacked=False)
    plt.title('Counts of Product Categories by Country', fontsize=16)
    plt.xlabel('Country', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='Product Category')
    plt.tight_layout()
    plt.show()

else:
    print("The required columns 'Country' and 'Product_Category' are not found in the dataset.")
````
